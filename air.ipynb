```python
import ast

# Define a function to convert string representation of dictionaries to actual dictionaries
def convert_to_dict(host_info):
    try:
        return ast.literal_eval(host_info)
    except:
        return None


data['host'] = data['host'].apply(convert_to_dict)

#new columns for the specific host details
data['host_id'] = data['host'].apply(lambda x: x.get('host_id') if x else None)
data['host_name'] = data['host'].apply(lambda x: x.get('host_name') if x else None)
data['host_location'] = data['host'].apply(lambda x: x.get('host_location') if x else None)
data['host_response_time'] = data['host'].apply(lambda x: x.get('host_response_time') if x else None)
data['host_response_rate'] = data['host'].apply(lambda x: x.get('host_response_rate') if x else None)
data['host_is_superhost'] = data['host'].apply(lambda x: x.get('host_is_superhost') if x else None)
data['host_listings_count'] = data['host'].apply(lambda x: x.get('host_listings_count') if x else None)
data['host_thumbnail_url'] = data['host'].apply(lambda x: x.get('host_thumbnail_url') if x else None)
```
